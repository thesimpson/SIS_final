```python
import numpy as np
import matplotlib.pyplot as plt
from random import *
import math
import numba
import scipy.stats
```

```python
num_sims = 100  # One Example
nf = float(num_sims)

dt = 0.02               # timestep Delta t
n = 10 # number of oscs
t_ini = 0               # starttime
t_end = 20.0           # endtime
n_stp = int(round((t_end-t_ini)/dt)) # step number

kappa, gamma, chi = 10,1,1
t = np.arange(t_ini, t_end + 0.5*dt, dt)

# np.savetxt("p.txt", p, fmt='%.18e', delimiter=',')
# p = np.loadtxt("p.txt", delimiter=",")
tmp_p = np.zeros(n)
for i in range(0,n): tmp_p[i] = 1 # random()
p = np.zeros((n,n))
for i in range(0,n):
    for j in range(0,n):
        p[i][j] = tmp_p[i]*tmp_p[j]
sum_p = np.sum(tmp_p)

s_in = np.zeros((n,3))
v_in = np.zeros((n,3))
s_in = np.loadtxt("s_in.txt",dtype=np.float64,delimiter=",")
v_in = np.loadtxt("v_in.txt",dtype=np.float64,delimiter=",")

# for i in range(0,n): # x[index][time][coord]
#     x_in[i][0] = uniform(-1,1)
#     x_in[i][1] = uniform(-1,1)
#     x_in[i][2] = uniform(-1,1)
#     s_in[i][0] = uniform(-1,1)
#     s_in[i][1] = uniform(-1,1)
#     s_in[i][2] = uniform(-1,1)
#     v_in[i][0] = ff(randint(0,1))
#     v_in[i][1] = ff(randint(0,1))
#     v_in[i][2] = ff(randint(0,1))
#     print(v_in[i])

#     v_in[i] /= np.linalg.norm(v_in[i])
#     s_in[i] = s_in[i] - (np.inner(v_in[i],s_in[i]))*v_in[i]
```

```python
@numba.njit

def dW(delta_t): # dW: Random process
    return np.random.normal(loc=0.0, scale=np.sqrt(delta_t))

def init(s_in,v_in):
    s = np.zeros((n,(n_stp + 1),3))
    v = np.zeros((n,(n_stp + 1),3))

    for i in range(0,n):
        s[i,0,:] = s_in[i]
        v[i,0,:] = v_in[i]
    return(s,v)

def IS_trial(sigma,s,v,model_num):
    for i in range(1, n_stp + 1): #iterate for time
        for j in range(0,n): # iterate for osc
            tmp_v1 = (sum_p/n) * np.average(v[:,i-1,:], axis=0, weights=tmp_p) # vc
            tmp_v2 = (kappa*tmp_p[j])*np.cross(v[j][i-1],tmp_v1) - (gamma/chi)*s[j][i-1] # coeff of dt
            if model_num==1:
                tmp_v3 = (sigma/chi) * s[j][i-1] # coeff of dw
                tmp_v4 = ((sigma/chi)**2) * s[j][i-1] #coeff of dw**2 - dt
            else:
                tmp_v3 = sigma*np.cross(v[j][i-1],tmp_v1)
                tmp_v4 = np.zeros(3)
            dw_ = dW(dt)
            v[j][i] = v[j][i-1] + (1/chi) * dt * np.cross(s[j][i-1],v[j][i-1])
            v[j][i] /= np.linalg.norm(v[j][i])

            s[j][i] = s[j][i-1] + dt*tmp_v2 + dw_*tmp_v3 + 0.5*(dw_**2-dt)*tmp_v4
    return(s,v)

def run_sims(s_in,v_in,sigma,num_sims,model_num):
    s1 = np.zeros((num_sims, n,(n_stp + 1),3))
    v1 = np.zeros((num_sims, n,(n_stp + 1),3))

    if sigma==0: 
        for ii in range(0,1):
            s1[ii],v1[ii] = init(s_in,v_in) # initialize
            s1[ii],v1[ii] = IS_trial(sigma,s1[ii],v1[ii],model_num)
        for ii in range(1,num_sims):
            s1[ii],v1[ii] = s1[ii-1],v1[ii-1]
    else:
        for ii in range(0,num_sims):
            s1[ii],v1[ii] = init(s_in,v_in) # initialize
            s1[ii],v1[ii] = IS_trial(sigma,s1[ii],v1[ii],model_num)
    return (s1,v1)
```

```python
sig = [2,3,4,5] # for model A
#sig = [0,2,3,5] # for model B

s1 = np.zeros((len(sig), num_sims, n,(n_stp + 1),3))
v1 = np.zeros((len(sig), num_sims, n,(n_stp + 1),3))

model_num = 1 # 1 for model A, otherwise model B
for i in range(0,len(sig)):
    s1[i], v1[i] = run_sims(s_in,v_in,sig[i],num_sims,model_num)

# s1 = np.reshape(np.loadtxt("s_m%d.csv" % model_num),[len(sig),num_sims,n,(n_stp + 1),3])
# v1 = np.reshape(np.loadtxt("v_m%d.csv" % model_num),[len(sig),num_sims,n,(n_stp + 1),3])
```

```python
s_norm = np.linalg.norm(s1, axis = 4) # s_norm = ((len(sig), num_sims, n,(n_stp + 1)))
St = np.mean(s_norm**2, axis = 2) # St = ((len(sig), num_sims, (n_stp + 1)))
St_mean = np.mean(St, axis = 1) # St_mean = ((len(sig), (n_stp + 1)))
St_stdv = np.sqrt((1/(num_sims-1))*np.var(St, axis = 1))

v_norm = np.linalg.norm(v1, axis = 4)**2 # v_norm2 = ((len(sig), num_sims, n,(n_stp + 1)))
vc = np.average(v1, axis = 2, weights = tmp_p) # ((len(sig), num_sims, (n_stp + 1),3))
Et = ((2*(sum_p**2))/(n**2))*(np.average(v_norm**2, axis=2, weights=tmp_p) - np.linalg.norm(vc, axis=3)**2) 
Et_mean = np.mean(Et, axis = 1) # Et_mean = ((len(sig), (n_stp + 1)))
Et_stdv = np.sqrt((1/(num_sims-1))*np.var(Et, axis = 1))

SE_plus = Et + (2/(chi*gamma))*St
SE_mean = np.mean(SE_plus, axis = 1)
SE_stdv = np.sqrt((1/(num_sims-1))*np.var(SE_plus, axis = 1))

```

```python
logSt = np.log10(St)
logSt_mean = np.mean(logSt, axis = 1)
logSt_stdv = np.sqrt((1/(num_sims-1))*np.var(logSt, axis = 1))

logEt = np.log10(Et)
logEt_mean = np.mean(logEt, axis = 1) # Et_mean = ((len(sig), (n_stp + 1)))
logEt_stdv = np.sqrt((1/(num_sims-1))*np.var(logEt, axis = 1))

logSE_plus = np.log10(SE_plus)
logSE_mean = np.mean(logSE_plus, axis = 1)
logSE_stdv = np.sqrt((1/(num_sims-1))*np.var(logSE_plus, axis = 1))
```
